### 3. Opening a csv file from an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/google/dspl/master/samples/google/canonical/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}

req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

### 9. Skiprows/ nrows Parameter

To skip particular rows use **skiprows** parameter.

```python
pd.read_csv('./DataSets/aug_train.csv')
```

```python
pd.read_csv('./DataSets/aug_train.csv', skiprows=[1,2,4])
#Row 1, 2, 4 will be skipped(0-indexed)
```

```python
pd.read_csv('./DataSets/aug_train.csv', skiprows=lambda x:(x%5)==0)
```

To take particular number of rows use **nrows** parameter

```python
pd.read_csv('./DataSets/aug_train.csv', nrows=100)
#Selecting first 100 rows
```

### 12. dtypes parameter

Using dtype parameter we can change datatype of a particular column.

**In the dataset the 'target' column can be represented as int but here it is float** (This occurs wastage of memory)

```python
pd.read_csv('./DataSets/aug_train.csv')
```

**Converting 'target' into int**

```python
pd.read_csv('./DataSets/aug_train.csv', dtype={'target':int})
```

### 13. Handling Dates

By default date is stored as object and that is difficult to manipulate. So **parse_date=[col-list]** will convert the datatype of date **object** to **datetime64**

```python
pd.read_csv('./DataSets/IPL Matches 2008-2020.csv')
```

```python
pd.read_csv('./DataSets/IPL Matches 2008-2020.csv').info()
```

```python
pd.read_csv('./DataSets/IPL Matches 2008-2020.csv', parse_dates=['date'])
```

```python
pd.read_csv('./DataSets/IPL Matches 2008-2020.csv', parse_dates=['date']).info()
```

### 14. Convertors

Using convertors we can convert any data by using function

```python
pd.read_csv('./DataSets/IPL Matches 2008-2020.csv')
```

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return name
```

```python
pd.read_csv('./DataSets/IPL Matches 2008-2020.csv', converters={'team1':rename})
```
